```python
html_page  = requests.get('https://incois.gov.in/MarineFisheries/MarineFisheryAdvisory.action/TextData?secid=SEC002#',headers=header).text
content = BeautifulSoup(html_page,'lxml')

print(content)

csv_table = content.find_all('table',class_='center')


csv_title = csv_table[len(csv_table)-1].find_all('th')
csv_rows = csv_table[len(csv_table)-1].find_all('tr')
dicti = {}
for title in csv_title:
    dicti[title.text]=[]


for row in csv_rows[1:]:

    cols = row.find_all('td')
    entry=[]
    for col in cols:
        entry.append(col.text)
    for key, value in zip(dicti.keys(), entry):
        dicti[key].append(value)



print(dicti)

csv_file = pd.DataFrame(dicti)
```

```python
df = csv_file
df
```

#Dataset Cleaning

```python
def dms_to_decimal(dms):
    parts = dms.split()
    degrees = int(parts[0])
    minutes = int(parts[1])
    seconds = int(parts[2])
    direction = parts[3]

    decimal_degrees = degrees + minutes / 60 + seconds / 3600
    if direction in ['S', 's', 'W', 'w']:
        decimal_degrees = -decimal_degrees

    return decimal_degrees
```

```python
for index, row in df.iterrows():
  df.iloc[index - 1, 5] = dms_to_decimal(df.iloc[index - 1, 5])
  df.iloc[index - 1, 6] = dms_to_decimal(df.iloc[index - 1, 6])
df
```

#API Call

```python
import requests
import json
```

```python
def get_weather_data(api):
        response = requests.get(f"{api}")
        if response.status_code == 200:
            return response.json()
        else:
            print(f"{response.status_code}")
```

```python
temp, max_temp, min_temp, pressure, humidity = [], [], [], [], []
for index, row in df.iterrows():
  data = get_weather_data(f'https://api.openweathermap.org/data/2.5/weather?lat={df.iloc[index - 1, 5]}&lon={df.iloc[index - 1, 6]}&appid=c1fe5cd49b5467294d552fa4f533122a')
  temp.append(data['main']['temp'])
  max_temp.append(data['main']['temp_max'])
  min_temp.append(data['main']['temp_min'])
  pressure.append(data['main']['pressure'])
  humidity.append(data['main']['humidity'])
```

```python
df['Temperature'] = temp
df['Max Temperature'] = max_temp
df['Min Temperature'] = min_temp
df['Pressure'] = pressure
df['Humidity'] = humidity
```
